# Functions

```python
def get_circumference(r):
    return math.pi * r * 2
```

```python
get_circumference(5)
```

```python
def binomial_coef(n,k):
    """
    This returns binomial coeff
    returns 
    n!/(k!*(n-k)!)
    """
    value = math.factorial(n)/(math.factorial(k) * math.factorial(n-k))
    return value

binomial_coef(52,2)
```
