# LINE PLOT

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define a data set
x= np.arange(0,10,0.1)
y= 2*x+5

#plot the data points
plt.plot(x,y)
plt.show()
```

# to customize the graph

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Changing the figure size
fig = plt.figure(figsize=(10,5))

# Define a data set
x= np.arange(0,10,1)
y= 2*x+5

#plot the data points
plt.plot(x,y,linewidth=2.0,linestyle=":",color="r",alpha=0.5,marker='o')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc="best")
plt.grid(True)
plt.show()
```

# Multiple Graphs

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Define a data set
x= np.arange(0,10,1)
y1= 2*x+5
y2= 3*x+10

#create 2 more plots now - in 1 figure
plt.subplot(3,2,1)
# try 121 top and 122 at bottom
#try same nos at top and bottom

#plotting points
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')



plt.show()
```
