# Box Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Prep
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order, discount])

#plotting the data
plt.boxplot(data, showmeans = True)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

# Violin Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Prep
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order, discount])

#plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Box Plot")
plt.grid(True)
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

label = ['Dog', 'Cat', 'Wolf', 'Lion']
sizes = [50,45,60,80]

plt.pie(sizes,labels = label,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title("Pie Chart")
plt.show()
```

# Doughnut Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

group_names = ['Group A', 'Group B', 'Group C']
group_size = [20,30,50]
size_center = [10]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]

pie1 = plt.pie(group_size, labels = group_names, radius = 1.5,colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_center, radius = 0.3, colors = 'w')
plt.show()
```

# How to create a Area Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = range(1,15)
y = [1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating area plot
plt.stackplot(x,y,color='Black',alpha=0.5)


#merge area chart with line chart
plt.plot(x,y,color='g')
plt.grid(True)
plt.show()
```
