# Model Building and Evaluation

## Build our Model 

**The model in this case requires a pipeline, because the input is text and it cannot be directly fed into our model.**

The pipeline will consist of a CountVectorizer --> TfidfTrasnformer --> Classifier.
Here we will use the entire train and test subsets of the original dataset

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn import metrics
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.datasets import fetch_20newsgroups
```

```python
X_train = twenty_train.data
X_test = twenty_test.data
y_train = twenty_train.target
y_test = twenty_test.target
```

```python
text_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('clf', KNeighborsClassifier(n_neighbors=7)),
                     ])

text_clf.fit(X_train, y_train)

predicted = text_clf.predict(X_test)

print(metrics.classification_report(y_test, predicted))
```

```python
for X_test, category in zip(X_test, predicted):
    print('%r => %s \n \n' % (X_test, twenty_train.target_names[category]))
```

```python
print("Test set score: {:.2f}".format(np.mean(predicted == y_test)))
```

We get a moderate accuracy using K means

```python
# Let's use svm classifier model to see i
```

```python
from sklearn.linear_model import SGDClassifier
svm_clf = Pipeline([
('vect', CountVectorizer()),
('tfidf', TfidfTransformer()),
('clf', SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, random_state=42,max_iter=5, tol=None)),])

svm_clf.fit(X_train, y_train)
```

```python
predicted = svm_clf.predict(twenty_test.data)
np.mean(predicted == twenty_test.target)
```

Using SV We got an accuracy of 83%, so compared to k means unable to get all in comparable groups,SVM performed better

```python
print(metrics.classification_report(y_test, predicted))
```
